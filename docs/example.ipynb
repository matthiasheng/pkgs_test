# Example Usage

Here we will demonstrate how to use pkgs_test to count words and plot them with some examples.

# Import package

```python
from pkgs_test.pkgs_test import count_words
from pkgs_test.plotting import plot_words
```

# Count words

The `count_words` function accept path to a text file and calculate number of words for each words inside the text, it removes the punctuations and make text in lower case.

```python
words_count = count_words('../tests/einstein.txt')
print(f'words_count: {words_count}')
```

# Plot words

The function `plot_words` accept the words counter object and plot the results in bar chart

```python
fig = plot_words(words_count)
```
